```python
import matplotlib.pyplot as plt
```

```python
study_time = [1, 5, 10, 20, 50, 80]
test_score_results = [40, 60, 70, 75, 88, 93]
```

```python
plt.title('Study Time vs Scores')
plt.xlabel('Study Time (minutes)')
plt.ylabel('Test Scores (out of 100)')
plt.plot(study_time, test_score_results)
plt.show()
```
